```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

fig, ax = plt.subplots(1, 3, figsize=(15, 6))

img0 = mpimg.imread("./beta=1e-6,iter=6e5.png")
img1 = mpimg.imread("./beta=5e-6,iter=6e5.png")
img2 = mpimg.imread("./beta=1.5e-6,iter=6e5.png")

images = [img0, img1, img2]

for i, img in enumerate(images):
    ax[i].imshow(img)
    ax[i].axis("off")

ax[0].set_title("beta=1e-6")
ax[1].set_title("beta=5e-6")
ax[2].set_title("beta=1.5e-5")

plt.show()
```
