```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

fig, ax = plt.subplots(2, 2, figsize=(8, 8))

img0 = mpimg.imread("./iter=0.png")
img1 = mpimg.imread("./beta=1e-8,iter=6e4.png")
img2 = mpimg.imread("./beta=5e-6,iter=6e4.png")
img3 = mpimg.imread("./beta=1.5e-6,iter=6e4.png")

images = [[img0, img2], [img1, img3]]

for i, imgX in enumerate(images):
    for j, img in enumerate(imgX):
        ax[j][i].imshow(img)
        ax[j][i].axis("off")

ax[0][0].set_title("iterations=0")
ax[1][0].set_title("beta=5e-6, iterations=60,000")
ax[0][1].set_title("beta=1e-8, iterations=60,000")
ax[1][1].set_title("beta=1.5e-5, iterations=60,000")

plt.show()
```
