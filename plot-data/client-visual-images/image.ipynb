```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

fig, ax = plt.subplots(1, 3, figsize=(12, 8))

img0 = mpimg.imread("./slice.png")
img1 = mpimg.imread("./3d.png")
img2 = mpimg.imread("./data.png")

images = [img0, img1, img2]

for i, img in enumerate(images):
    ax[i].imshow(img)
    ax[i].axis("off")

ax[0].set_title("Slice view")
ax[1].set_title("3D view")
ax[2].set_title("Data view")

plt.show()
```
