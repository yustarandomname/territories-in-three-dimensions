```python
import matplotlib.image as mpimg
import numpy as np

# Lattice size = 20
# iterations = 50 000
# Total mass per species = 1_600_000
# Beta = Variable
# Gamma = Variable
# Lambda = 0.5

gamma_with_approx_critical_beta = {
    "0.04": {
        "over": {"beta": 5e-5, "order_param": 0.01251499917749875},
        "under": {"beta": 4e-5, "order_param": 0.007640832101696207},
    },
    "0.0625": {
        "over": {"beta": 3e-5, "order_param": 0.010533777871662375},
        "under": {"beta": 2.7e-5, "order_param": 0.008266412812644744},
    },
    "0.125": {
        "over": {"beta": 1.5e-5, "order_param": 0.010533777871662375},
        "under": {"beta": 1.3e-5, "order_param": 0.007838046943427157},
    },
    "0.25": {
        "over": {"beta": 8e-6, "order_param": 0.012373760883934354},
        "under": {"beta": 6e-6, "order_param": 0.0068595529592293675},
    },
    "0.5": {
        "over": {"beta": 4e-6, "order_param": 0.012373760883934354},
        "under": {"beta": 2e-6, "order_param": 0.00449186733302237},
    },
    "0.75": {
        "over": {"beta": 3e-6, "order_param": 0.01816291594700141},
        "under": {"beta": 2e-6, "order_param": 0.006853646181443476},
    },
    "1.0": {
        "over": {"beta": 2e-6, "order_param": 0.012373760883934354},
        "under": {"beta": 1.5e-6, "order_param": 0.0068595529592293675},
    },
    "1.5": {
        "over": {"beta": 1.4e-6, "order_param": 0.014484587662608683},
        "under": {"beta": 1.2e-6, "order_param": 0.00962262997823643},
    },
    "2.0": {
        "over": {"beta": 1.2e-6, "order_param": 0.023805522073824525},
        "under": {"beta": 8e-7, "order_param": 0.007619285489664465},
    },
}

target = 0.01

fig, ax = plt.subplots(1, 2, figsize=(12, 4))


# The beta value that would correspond to the target order parameter is
# linearly interpolated between the two closest order parameters values
def lerp_beta(x):
    over = x["over"]
    under = x["under"]

    delta_over = over["order_param"] - target
    delta_under = target - under["order_param"]
    delta_sum = delta_over + delta_under
    return (
        over["beta"] * delta_under / delta_sum + under["beta"] * delta_over / delta_sum
    )


gamma_critical_betas = map(
    lambda x: lerp_beta(x), gamma_with_approx_critical_beta.values()
)
gammas = list(
    map(lambda x: float(x) / float(0.5), gamma_with_approx_critical_beta.keys())
)
critical_betas = list(gamma_critical_betas)
print(list(zip(gammas, critical_betas)))

ax[1].title.set_text(
    "  Critical Beta vs $\gamma$/$\lambda$ 3D Discreet Model \n Mass = 3200000  "
)
ax[1].plot(gammas, critical_betas)
ax[1].scatter(gammas, critical_betas)
ax[1].set_xlabel("$\gamma$ / $\lambda$")
ax[1].set_ylabel("$\\beta$*")
ax[1].set_ylim(0, 5e-5)


img = mpimg.imread("./gamma-v-lambda-2d.png")
ax[0].imshow(img)
ax[0].axis("off")

# x_cords = range(2_000_000, 45_000_000, 100_000)
# y_cords = [1 / x for x in x_cords]

ax[0].title.set_text("  Critical Beta vs $\gamma$/$\lambda$ 2D \n Mass = 200000  ")
# ax[0].plot(x_cords, y_cords, color="red")
# ax[0].set_xlabel("Gamma / Lambda")
# ax[0].set_ylabel("Critical Beta")

plt.show()
```
