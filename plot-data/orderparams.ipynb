```python
import pandas as pd
from matplotlib import pyplot as plt
plt.rcParams["figure.autolayout"] = True
columns = ["iter", "result"]

betas = [1e-5,5e-6,1e-6]
fig, ax = plt.subplots()

for beta in betas:
    df = pd.read_csv("data/beta="+str(beta)+".csv", usecols=columns)
    ax.plot(df.iter, df.result, label="beta="+str(beta))

ax.legend(loc='upper left', bbox_to_anchor=(1.01, 1.01))

plt.title("Order Parameter over iterations with different beta values")
plt.ylabel("Order Parameter")
plt.xlabel("Iterations")
plt.tight_layout()
plt.show()
```
