```python
# Lattice size = 20
# iterations = 50 000
# Total mass per species = Variable
# Beta = Variable
# Gamma = 0.5
# Lambda = 0.5

gamma_with_approx_critical_beta = {
    "1_000_000": {
        "over": {"beta": 5e-6, "order_param": 0.010605389290576306},
        "under": {"beta": 4.4e-6, "order_param": 0.00900553530514639},
    },
    "1_200_000": {
        "over": {"beta": 4.5e-6, "order_param": 0.010030701448545232},
        "under": {"beta": 4.4e-6, "order_param": 0.009613688500768491},
    },
    "1_600_000": {
        "over": {"beta": 4e-6, "order_param": 0.012465702550012447},
        "under": {"beta": 3e-6, "order_param": 0.006877975353287991},
    },
    "2_400_000": {
        "over": {"beta": 3e-6, "order_param": 0.012789843255848254},
        "under": {"beta": 2.5e-6, "order_param": 0.007214229108326337},
    },
    "3_200_000": {
        "over": {"beta": 2.5e-6, "order_param": 0.012934357165436547},
        "under": {"beta": 2.2e-6, "order_param": 0.007807194535832243},
    },
    "6_400_000": {
        "over": {"beta": 1.4e-6, "order_param": 0.01621520857575315},
        "under": {"beta": 1.2e-6, "order_param": 0.005791238254117009},
    },
    "12_800_000": {
        "over": {"beta": 7.1e-7, "order_param": 0.01264055732899372},
        "under": {"beta": 6.9e-7, "order_param": 0.009059955465463544},
    },
    "19_200_000": {
        "over": {"beta": 5e-7, "order_param": 0.01648507518175625},
        "under": {"beta": 4.8e-7, "order_param": 0.008675341075132539},
    },
    "22_400_000": {
        "over": {"beta": 4.4e-7, "order_param": 0.025272879107708635},
        "under": {"beta": 4.2e-7, "order_param": 0.0086422342830665},
    },
}

target = 0.01

fig, ax = plt.subplots(1, 2, figsize=(12, 4))


# The beta value that would correspond to the target order parameter is
# linearly interpolated between the two closest order parameters values
def lerp_beta(x):
    over = x["over"]
    under = x["under"]

    delta_over = over["order_param"] - target
    delta_under = target - under["order_param"]
    delta_sum = delta_over + delta_under
    return (
        over["beta"] * delta_under / delta_sum + under["beta"] * delta_over / delta_sum
    )


mass_critical_betas = map(
    lambda x: lerp_beta(x), gamma_with_approx_critical_beta.values()
)
total_mass = list(map(lambda x: int(x) * 2, gamma_with_approx_critical_beta.keys()))
critical_betas = list(mass_critical_betas)
print(list(critical_betas))

ax[1].title.set_text("  Critical Beta vs Mass 3D Discreet Model  ")
ax[1].plot(total_mass, critical_betas)
ax[1].scatter(total_mass, critical_betas)
ax[1].set_xlabel("Mass")
ax[1].set_ylabel("Critical Beta")

x_cords = range(4_000_000, 45_000_000, 100_000)
y_cords = [1 / x for x in x_cords]

ax[0].title.set_text("Critical Beta vs Mass 2D Stability Analysis")
ax[0].plot(x_cords, y_cords, color="red")
ax[0].set_xlabel("Mass")
ax[0].set_ylabel("Critical Beta")

plt.show()
```
